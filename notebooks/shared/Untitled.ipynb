```python
import pandas as pd
#make dataframe and remove index column
df = pd.read_csv(r'C:/Users/aminp/OneDrive/Dokumenter/NTNU/4Ã¥r/Machinelearning/machine_learning/data/train_data/dataset_train_2.csv', index_col=0)
df=df.drop('store_id',axis=1)
```

In the next cell we import the sklearn library and begin predicting. I am only using the test set to fit and predict. We impude the values that are nan, and check with two different kernels to see what models fits best. 

```python
import sklearn
import numpy as np
from sklearn import svm, preprocessing
from sklearn.impute import SimpleImputer

df = sklearn.utils.shuffle(df)

X=df.drop('revenue',axis=1).values
X=preprocessing.scale(X)
y=df['revenue'].values

test_size = 200

X_train = X[:-test_size]
y_train = y[:-test_size]

X_test = X[-test_size:]
y_test = y[-test_size:]

#impuding nan-values with mean, just to make the code runnable
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(X_train)
X_train_imp = imp.transform(X_train)
X_test_imp = imp.transform(X_test)

clf = svm.SVR(kernel='linear')
clf.fit(X_train_imp,y_train)
```

```python
clf.score(X_test_imp,y_test)
```

```python
clf = svm.SVR(kernel='rbf')
clf.fit(X_train_imp,y_train)
```

```python
clf.score(X_test_imp,y_test)
```

```python
for X,y in zip(X_test_imp,y_test):
    print(f"Predicted: {clf.predict([X])[0]},   Actual: {y}")
```
